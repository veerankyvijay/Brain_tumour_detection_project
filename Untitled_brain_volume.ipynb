```python
import os
import numpy as np
import nibabel as nib
import matplotlib.pyplot as plt
from scipy.ndimage import label, center_of_mass
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle
import tensorflow as tf

# Define file paths
mri_path = r"C:/Users/vijay/Desktop/brain_tumour_segmentation_project/MRI"
ultrasound_path = r"C:/Users/vijay/Desktop/brain_tumour_segmentation_project/Ultrasound"

# Define a function to load NIfTI files and preprocess them
def load_nii_file(file_path):
    nii_image = nib.load(file_path)
    data = nii_image.get_fdata()
    return data

# Define a function to calculate tumor volume
def calculate_volume(data, voxel_size):
    # Threshold to identify tumor region
    binary_mask = data > 0.5  # Adjust threshold based on dataset
    labeled_array, num_features = label(binary_mask)
    tumor_volume = np.sum(binary_mask) * np.prod(voxel_size)
    return tumor_volume

# Load MRI and Ultrasound files
mri_files = [os.path.join(mri_path, f) for f in os.listdir(mri_path) if f.endswith('.nii')]
ultrasound_files = [os.path.join(ultrasound_path, f) for f in os.listdir(ultrasound_path) if f.endswith('.nii')]

# Example: Process first MRI file
if mri_files:
    mri_data = load_nii_file(mri_files[1])
    voxel_size = nib.load(mri_files[1]).header.get_zooms()  # Extract voxel size
    tumor_volume = calculate_volume(mri_data, voxel_size)
    print(f"Tumor Volume for {mri_files[1]}: {tumor_volume:.2f} mm^3")

# Visualization of MRI slices
    plt.figure(figsize=(12, 6))
    slice_idx = mri_data.shape[2] // 2  # Middle slice
    plt.imshow(mri_data[:, :, slice_idx], cmap='gray')
    plt.title('MRI Slice')
    plt.colorbar()
    plt.show()

# Build a CNN model for volume prediction
model = Sequential()
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(150, 150, 3)))
model.add(MaxPooling2D(2, 2))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(2, 2))
model.add(Dropout(0.3))
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D(2, 2))
model.add(Dropout(0.3))
model.add(Flatten())
model.add(Dense(512, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(1, activation='linear'))  # Output layer for volume prediction

model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mae'])
model.summary()

# Prepare data for training (example dataset loading placeholder)
# Assuming data (X_train, y_train) and labels (tumor volumes)
X_train, X_test, y_train, y_test = train_test_split(np.random.rand(100, 150, 150, 3), 
                                                    np.random.rand(100, 1), 
                                                    test_size=0.1, 
                                                    random_state=101)

# Train the model
history = model.fit(X_train, y_train, epochs=20, validation_split=0.1, batch_size=16)

model.save_weights('Tumour_volume.weights.h5')
model_json = model.to_json()
with open("tumor_volume_model.json", "w") as json_file:
    json_file.write(model_json)

# Plot training and validation metrics
plt.figure(figsize=(14, 7))
plt.plot(history.history['mae'], label='Training MAE')
plt.plot(history.history['val_mae'], label='Validation MAE')
plt.title('Training and Validation Mean Absolute Error')
plt.legend()
plt.show()
```

```python
# Define a function to preprocess MRI/Ultrasound data (resize and normalize)
from tensorflow.keras.preprocessing import image
import numpy as np

def preprocess_data(file_path, target_size=(150, 150)):
    data = load_nii_file(file_path)
    slice_idx = data.shape[2] // 2  # Take the middle slice for prediction
    slice_image = data[:, :, slice_idx]  # Extract the middle slice
    slice_image_resized = np.resize(slice_image, target_size + (1,))  # Resize to target size (150, 150)
    slice_image_resized = np.repeat(slice_image_resized, 3, axis=-1)  # Repeat to match 3 channels (RGB)
    slice_image_resized = np.expand_dims(slice_image_resized, axis=0)  # Add batch dimension
    return slice_image_resized

# Load model weights and structure from the saved files
from keras.models import model_from_json

# Load the model structure from JSON
with open('tumor_volume_model.json', 'r') as json_file:
    model_json = json_file.read()

# Rebuild the model from JSON
model = model_from_json(model_json)

# Load model weights
model.load_weights('Tumour_volume.weights.h5')

# Specify the path to the MRI/Ultrasound file you want to predict tumor volume for
input_file = r"C:/Users/vijay/Desktop/brain_tumour_segmentation_project/MRI/cropped_MRI1.nii"  # Replace with your file path

# Preprocess the data
input_data = preprocess_data(input_file)

# Predict tumor volume
predicted_volume = model.predict(input_data)

# Print the predicted tumor volume
print(f"Predicted Tumor Volume: {predicted_volume[0][0]:.2f} mm^3")
```
